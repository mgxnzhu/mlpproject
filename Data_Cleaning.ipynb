```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("creditcard.csv")
data = (data-data.min())/(data.max()-data.min())
pos_data = data[data["Class"]==1]
neg_data = data[data["Class"]==0]
for key in data.keys()[1:-2]: # for all principle components
    fig = plt.figure(figsize=(4,3))
    plt.hist(pos_data[key], histtype="step", normed=True, bins=50, range=(0,1), label="pos")
    plt.hist(neg_data[key], histtype="step", normed=True, bins=50, range=(0,1), label="neg")
    plt.xlim(0,1)
    plt.title(key)
    plt.legend()
    fig.savefig(key+"_Scaled.png")
    
stats = pd.DataFrame({'mean':data.mean(), 'std':data.std(), 'skew':data.skew()})
```
